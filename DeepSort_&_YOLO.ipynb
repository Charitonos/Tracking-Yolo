```python
import torch

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

# Select the detection model
model = models.get("yolo_nas_l" , pretrained_weights="coco").to(device)

# Switch from training to evaluation or inference
model.eval()
```

```python
# get frame info for tracker and video saving
h,w = (720,1280)
h2,w2 = h // 2 , w // 2

img_info = (h,w)
img_size = (h2,w2)

# This function is handling the dection preedictions
# We provide the detection we got using YOLO-NAS
# It will preprocess that input. Then convert it to a torch tensor.The torch
# tensor will become the input to the model. Lastly it will decode the output
def get_prediction(image_in, pipeline):
    # '''' Obtains DetectionPrediction object from a single input RGB image''''
    # Preprocess
    preprocessed_image, processing_metadata = pipeline.image_processor.preprocess_image(image = image_in.copy())

    # Predict
    with torch.no_grad():
        torch_input = torch.Tensor(preprocessed_image).unsqueeze(0).to(device)
        model_output = model(torch_input)
        prediction = pipeline._decode_model_output(model_output , model_input = torch_input)

    # Postprocess
    return pipeline.image_processor.postprocess_predictions(predictions = prediction[0], metadata=processing_metadata)
```

```python
pipeline = DetectionPipeline(
    model=model,
    image_processor=model._image_processor,
    post_prediction_callback=model.get_post_prediction_callback(iou=0.55, conf=0.2),
    class_names=model._class_names,
)
```

#Running the algorithm

Run the code segment below in order to get an output of tracking objects in a video

```python
import os
import random
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
from ultralytics import YOLO
from tracker import Tracker

# Set up the path of the video.
# input path
video_path = os.path.join('.' , 'data' , 'Mall.mkv')
# output path
video_path_out = os.path.join('.' , 'data' , 'out.mp4')

# Capture the video from the video path.
cap = cv.VideoCapture(video_path)

# Read the first frame.
# ret returns true if the frame is succesfully read.
# frame contains the frame that was read.
ret , frame = cap.read()

# Capture the output
cap_out = cv.VideoWriter(video_path_out , cv.VideoWriter_fourcc(*'MP4V'), cap.get(cv.CAP_PROP_FPS),
                          (frame.shape[1], frame.shape[0]))

# initialize tracker
tracker = Tracker()

# Random colors for bounding boxes.
colors = [((random.randint(0,255)) , (random.randint(0,255)) , (random.randint(0,255))) for j in range(20)]

# Set the conf here
conf = 0.2

while ret:

    # If the model is set to Yolov8 use this.
    if yolo_model == "v8":
        results = model(frame)
        for result in results:
            detections = []
            labels = []
            for r in result.boxes.data.tolist():
                x1 , y1 , x2 , y2 , pc , class_id = r
                x1 = int(x1)
                y1 = int(y1)
                x2 = int(x2)
                y2 = int(y2)
                class_id = int(class_id)
                if pc > conf and class_id in CLASS_ID:
                    detections.append([x1, y1, x2, y2, pc])
                    labels.append(class_id)

    # If the model is set to Yolo-NAS use this.
    if yolo_model == "NAS":
        results = get_prediction(frame , pipeline)
        detections = []
        labels = []
        i = 0

        # Results from predictions
        xyxys = results.bboxes_xyxy
        confidences = results.confidence
        class_ids = results.labels.astype(int)

        #print(confidences)

        for bbox in xyxys:
            x1 , y1 , x2 , y2 = bbox
            x1 = int(x1)
            y1 = int(y1)
            x2 = int(x2)
            y2 = int(y2)
            class_id = int(class_ids[i])
            pc = confidences[i]
            i = i + 1
            if pc > conf and class_id in CLASS_ID:
                detections.append([x1, y1, x2, y2, pc])
                labels.append(class_id)


    tracker.update(frame , detections)

    # Each trck is an object in the video.
    for track in tracker.tracks:

        # We are accesing the bounding box of the object.
        bbox = track.bbox
        x1 , y1 , x2 , y2 = bbox

        # Accesing the id DeepSort assigned to this object.
        track_id = track.track_id

        # Assign a color to different people. Draw the rectamgles on the frame.
        cv.rectangle(frame, (int(x1) , int(y1)) , (int(x2) , int(y2)) , (255, 0, 0) , 2)

        # Assign the label to the bounding box of the object
        cv.putText(frame , f"ID: {str(track.track_id)}", (int(x1), int(y1) - 10), fontFace = cv.FONT_HERSHEY_SIMPLEX,
                    fontScale = 0.6, color = (255, 255, 255), thickness=2 )

    # Write the current frame
    cap_out.write(frame)
    # Read the next frame
    ret , frame = cap.read()

# Free memory.
cap.release()
cap_out.release()
cv.destroyAllWindows()
```
